```python
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l    
```

```python
import ast
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords']=movies['keywords'].apply(convert)#now keywords are aranged
```

```python
movies.head()
```

```python
def convert3(obj):
    l=[]
    counter =0
    for i in ast.literal_eval(obj):
        if counter !=3:
         l.append(i['name'])
         counter+=1
        else:
         break
    return l  
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
def fetch_director(obj):
    l=[]
    for i in ast.literal_eval(obj):
       if i['job']=='Director':
            l.append(i['name'])
            break         
    return l  
```

```python
movies['crew'][0]
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
#making director nbame only in creq
```

```python
#making list in overview
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
#remove spaces in words
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head()
```

```python
movies['tags']=movies['overview']+movies['genres']+ movies['keywords']+movies['cast']+movies['crew']
```

```python
new= movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
new['tags']=new['tags'].apply(lambda x:" ".join(x))
new.head()
```

```python
new.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector =cv.fit_transform(new['tags']).toarray()
```

```python
vector[0]
```

```python
#we ll apply stemming which makes alll similar word into one like - loved,loving, loves will be love
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
new[new['title'] == 'The Lego Movie'].index[0]
```

```python
def recommend(movie):
    index=new[new['title'] == movie].index[0]
    distance = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distance[1:6]:
        print(new.iloc[i[0]].title)
```

```python
recommend('Life Of Pie')
```
